```python
# Импортируем необходимые для работы библиотеки
import pandas as pd
import numpy as np
import random
import matplotlib
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
# Импортируем данные
df = pd.read_csv('beverage_r.csv', sep=";", index_col='numb.obs')
df.head()
```

```python
len(df)
```

**Задача 1.**

Разделить таблицу данных так, чтобы  в обучающую выборку попало 66% наблюдений, а в тестовую выборку 33% наблюдений

```python
#  Зерно датчика случайных чисел
np.random.seed(seed=123456)
# Генерируем случайные числа от 0.0 до 1.0
r1 = np.random.uniform(low = 0.0, high=1.0, size=len(df))

# Для точного разделения используем квантили
test = r1 > np.quantile(r1, 0.66)
train = np.logical_not(r1 > np.quantile(r1, 0.66))

df_test = df[test]
df_train = df[train]

print(f'Тестовая выбора: {len(df_test)} наблюдений')
print(f'Обучающая выбора: {len(df_train)} наблюдения')

```

**Задача 2.**

Разделить таблицу данных на четыре одинаковых части.

```python
#  Зерно датчика случайных чисел
np.random.seed(seed=123456)
# Генерируем случайные числа от 0.0 до 1.0
r1 = np.random.uniform(low = 0.0, high=1.0, size=len(df))

# Для точного разделения используем квантили
first = r1 <= np.quantile(r1, 0.25)
second = (r1 > np.quantile(r1, 0.25)) & (r1 <= np.quantile(r1, 0.50))
third = (r1 > np.quantile(r1, 0.50)) & (r1 <= np.quantile(r1, 0.75))
fourth = r1 > np.quantile(r1, 0.75)

df_first = df[first]
df_second = df[second]
df_third = df[third]
df_fourth = df[fourth]

print(f'Первая часть выбори: {len(df_first)} наблюдений')
print(f'Вторая часть выборки: {len(df_second)} наблюдений')
print(f'Третья часть выборки: {len(df_third)} наблюдений')
print(f'Четвертая часть выборки: {len(df_fourth)} наблюдений')
```

**Задача 3.**

Разделить таблицу данных на две стратифицированные подвыборки.
По переменным кока-кола и минералка.

*(в представленном датасете нет переменной "минералка", поэтому она будет заменена на переменную D_COKE)*

```python
# Создаем переменную strat
strat = df['COKE'].astype(str) + '_' + df['D_COKE'].astype(str)

# Разделяем данные на две выборки
df_train, df_test = train_test_split(df, test_size=0.33, stratify=strat, random_state=123456)

print(f'Исходная выбора: {len(df)} наблюдения')
print('Исходное распределение переменных')
print(df[['COKE','D_COKE']].value_counts())

print(f'\nТестовая выбора: {len(df_test[['COKE','D_COKE']])} наблюдений')
print('Распределение переменных в тестовой выборке')
print(df_test[['COKE','D_COKE']].value_counts())

print(f'\nОбучающая выбора: {len(df_train[['COKE','D_COKE']])} наблюдения')
print('Распределение переменных в обучающей выборке')
print(df_train[['COKE','D_COKE']].value_counts())
```

Можно отметить, что стратификация работает корректно, в обеих выборках представлены все комбинации, а также сохранены пропорции.
